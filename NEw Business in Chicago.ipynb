#### Send the GET Request and examine the results

```python
results = requests.get(url).json()
results
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
df_chicago= json_normalize(venues)
df_chicago.head()
```

#### Define information of interest and filter dataframe

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in df_chicago.columns if col.startswith('location.')] + ['id']
df_chicago_filtered = df_chicago.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
df_chicago_filtered['categories'] = df_chicago_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
df_chicago_filtered.columns = [column.split('.')[-1] for column in df_chicago_filtered.columns]

df_chicago_filtered
```

```python
df_chicago_filtered.name
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=13) # generate map centred around the Conrad Hotel

# add a red circle marker to represent the Conrad Hotel
folium.features.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='Chicago',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)

# add the Italian restaurants as blue circle markers
for lat, lng, label in zip(df_chicago_filtered.lat, df_chicago_filtered.lng, df_chicago_filtered.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)

# display map
venues_map
```
